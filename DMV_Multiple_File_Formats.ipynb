```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
iris_data = pd.read_json('Datasets/DMV/iris.json')
iris_data.head()
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepalLength', data=iris_data)
plt.title('Sepal Length by Species in Iris Dataset')
plt.show()
```

```
titanic_data = pd.read_excel('Datasets/DMV/Titanic.xlsx')
titanic_data.head()
```

```
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Survival Count by Passenger Class in Titanic Dataset')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```
boston_data = pd.read_csv('Datasets/DMV/boston.csv')
boston_data.head()
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RM', y='MEDV', data=boston_data)
plt.title('Median Value of Homes vs. Average Number of Rooms in Boston Housing Dataset')
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median Value of Homes (MEDV)')
plt.show()
```
