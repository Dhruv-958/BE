```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```
df = pd.read_csv('Wine.csv')
```

```
X = df.drop('Customer_Segment', axis=1)
y = df['Customer_Segment']
```

```
plt.figure(figsize=(10,6))
plt.scatter(X[y == 1]['Alcohol'], X[y == 1]['Malic_Acid'], label = 'CS_1', alpha=0.7)
plt.scatter(X[y == 2]['Alcohol'], X[y == 2]['Malic_Acid'], label = 'CS_2', alpha=0.7)
plt.scatter(X[y == 3]['Alcohol'], X[y == 3]['Malic_Acid'], label = 'CS_3', alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.legend()
plt.show()
```

```
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Customer_Segment'] = y
```

```
plt.figure(figsize=(10,6))
plt.scatter(pca_df[pca_df['Customer_Segment'] == 1]['PC1'], pca_df[pca_df['Customer_Segment'] == 1]['PC2'], label = 'CS_1', alpha=0.7)
plt.scatter(pca_df[pca_df['Customer_Segment'] == 2]['PC1'], pca_df[pca_df['Customer_Segment'] == 2]['PC2'], label = 'CS_1', alpha=0.7)
plt.scatter(pca_df[pca_df['Customer_Segment'] == 3]['PC1'], pca_df[pca_df['Customer_Segment'] == 3]['PC2'], label = 'CS_1', alpha=0.7)
```
