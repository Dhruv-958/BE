```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
df = pd.read_csv('Telecom_Customer_Churn.csv')
df
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
categorical_columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
                       'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for column in categorical_columns:
    if column in df.columns:
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Ensure numerical columns are actually numeric
numerical_columns = ['MonthlyCharges', 'TotalCharges', 'tenure']

for column in numerical_columns:
    if column in df.columns:
        # Convert column to numeric, coercing errors to NaN
        df[column] = pd.to_numeric(df[column], errors='coerce')

# Now fill missing values with the median of each column
for column in numerical_columns:
    if column in df.columns:
        df[column].fillna(df[column].median(), inplace=True)
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns
z_scores = np.abs(stats.zscore(df[numeric_columns]))
df = df[(z_scores < 3).all(axis=1)]
```

```python
scaler = StandardScaler()
numerical_features = ['MonthlyCharges', 'TotalCharges', 'tenure']  # Assuming these columns are in the dataset
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
# Convert target variable 'Churn' to binary values (1 for churn, 0 for no churn)
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
# Dropping irrelevant columns and preparing the feature set and target variable
X = df.drop(['customerID', 'Churn'], axis=1)
y = df['Churn']
```

```python
# Export the cleaned dataset
df.to_csv('Cleaned_Telecom_Customer_Churn.csv', index=False)
df
```
